```python
import sys
import os
import numpy as np 
import pandas as pd
from pathlib import Path

# univariate mlp example
from numpy import array

from keras import optimizers
from keras import regularizers
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Conv1D, MaxPooling1D, Flatten, Dropout, LSTM

from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.externals import joblib 

import matplotlib.pyplot as plt
```

```python
PATH = Path("../data/")
```

```python
scaler = joblib.load(PATH/'train_scaler_x.pkl')
```

```python
dfX_Train = np.load(PATH/'train_x.npy')
dfX_Val = np.load(PATH/'val_x.npy') 
dfX_Test = np.load(PATH/'test_x.npy')
dfY_Train = np.load(PATH/'train_y.npy')
dfY_Val = np.load(PATH/'val_y.npy')
dfY_Test = np.load(PATH/'test_y.npy')
print(dfX_Train.shape, dfY_Train.shape)
print(dfX_Val.shape, dfY_Val.shape)
print(dfX_Test.shape, dfY_Test.shape)
print(dfX_Train[0,:10])
```

```python
plt.plot(dfX_Train[0,:1250])
plt.show()
plt.plot(dfX_Train[0,1250:])
plt.show()
```

```python
dim = 125*5
dfY_Tr = dfY_Train[:,:dim]
dfY_Va = dfY_Val[:,:dim]
dfY_Te = dfY_Test[:,:dim]

scaler_0 = MinMaxScaler().fit(dfX_Train[:,:dim])
scaler_1 = MinMaxScaler().fit(dfX_Train[:,1250:(1250+(dim))])
dfX_Tr_0 = scaler_0.transform(dfX_Train[:,:dim])
dfX_Tr_1 = scaler_1.transform(dfX_Train[:,1250:(1250+(dim))])

dfX_Va_0 = scaler_0.transform(dfX_Val[:,:dim])
dfX_Va_1 = scaler_1.transform(dfX_Val[:,1250:(1250+(dim))])

dfX_Te_0 = scaler_0.transform(dfX_Test[:,:dim])
dfX_Te_1 = scaler_1.transform(dfX_Test[:,1250:(1250+(dim))])                                        
```

```python
trainX = np.stack((dfX_Tr_0[:,:], dfX_Tr_1[:,:]), axis=2)
valX = np.stack((dfX_Va_0[:,:], dfX_Va_1[:,:]), axis=2)
testX = np.stack((dfX_Te_0[:,:], dfX_Te_1[:,:]), axis=2)

print(trainX.shape, valX.shape, testX.shape)
```

```python
ix = 50
plt.plot(trainX[ix,:,0])
plt.show()
plt.plot(trainX[ix,:,1])
plt.show()
plt.plot(dfY_Te[ix,:])
plt.show()
```

```python
bs=100
look_back = 125*5
model = Sequential()
model.add(LSTM(150, batch_input_shape=(bs,look_back,2),dropout=0.4, stateful=True, return_sequences=True))
model.add(LSTM(150, batch_input_shape=(bs,look_back,2),dropout=0.4, stateful=True))
model.add(Dense(125*5))
model.compile(loss='mean_squared_error', optimizer=optimizers.Adam(lr=1e-2))
model.summary()
```

```python
# fit model
model.compile(optimizer=optimizers.Adam(lr=1e-2), loss='mse')
for i in range(5):
    history = model.fit(trainX[:8000], dfY_Tr[:8000], validation_data=(valX[:1000], dfY_Va[:1000]), 
                        batch_size= bs, epochs=1, verbose=1)
    model.reset_states()
```

```python
# fit model
model.compile(optimizer=optimizers.Adam(lr=1e-3), loss='mse')
for i in range(5):
    history = model.fit(trainX[:8000], dfY_Tr[:8000], validation_data=(valX[:1000], dfY_Va[:1000]), 
                        batch_size= bs, epochs=1, verbose=1)
    model.reset_states()
```

```python
# fit model
model.compile(optimizer=optimizers.Adam(lr=1e-4), loss='mse')
for i in range(5):
    history = model.fit(trainX[:8000], dfY_Tr[:8000], validation_data=(valX[:1000], dfY_Va[:1000]), 
                        batch_size= bs, epochs=1, verbose=1)
    model.reset_states()
```

```python
# plot history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
